```python
## Q6
import numpy as np
### get median
medBef = df['horsepower'].median()
print(f'Meadian before: {medBef}')

### most frequent value
mfv = df['horsepower'].value_counts().reset_index()

### most frequent value
mostFreVal = mfv['count'].max()

### fill the missing values
df['horsepower'].fillna(value = mostFreVal, inplace = True)

### get median
medAft = df['horsepower'].median()
print(f'Meadian after: {medAft}')
```
